```python
# 获得smiles列表
smiles_list=np.array(df['smiles'])
smiles_list
```

```python
from rdkit import Chem
```

```python
# 获得每个分子包含的原子数
N=[]  # 每个分子包含的原子个数
for smiles in smiles_list:
    mol = Chem.MolFromSmiles(smiles)
    atom_num = mol.GetNumAtoms()
    N.append(atom_num)
N=np.array(N)
print(N)
N.shape
```

```python
# 下一步获得R
# 遍历mol文件夹中的文件
path = 'E:/马金博/缓蚀剂研究/三维数据集制作/test3/'
file_txt = os.listdir(path)
file_txt.sort(key=lambda x:int(x[:-4]))  #解决os.listdir读取数据的乱序问题
# print(file_txt)
contents=[]
for n,name in zip(N,file_txt):
    file_path=os.path.join(path, name)
#     print(file_path)
    f = open(file_path) # 依次读取文件
#     print(f)
    content=f.read().split('\n')
#     print(name)
    for i in range(4,n+4):
#         print(i)
        a=content[i].split()
        contents.append(a[0:4])
       
    
print(contents)        
    
# lines = f.read() 
# lines
# a=lines.split('\n')
# a[5].split()
# # for line in lines:
# #     print(line)
```

```python
# 获得原子
atoms=[]
for content in contents:
    atom=content[3]
    atoms.append(atom)
print(atoms)
```

```python
# 获得原子对应得坐标，并转为数字
R=[]
for content in contents:
    r_float=[]
    for i in content[:3]:
        r_float.append(float(i))
    R.append(r_float)
R=np.array(R)
R
```

```python
# 计算对应原子的核电荷数
Z = []
for atom in atoms:
    if atom=='N':
        Z.append(7)
    elif atom=='Br':
        Z.append(35)
    elif atom=='C':
        Z.append(6)
    elif atom=='O':
        Z.append(8)
    elif atom=='K':
        Z.append(19)
    elif atom=='S':
        Z.append(16)
    elif atom=='Cl':
        Z.append(17)
    elif atom=='F':
        Z.append(9)
    elif atom=='P':
        Z.append(15)
    elif atom=='H':
        Z.append(1)
    elif atom=='I':
        Z.append(53)
    elif atom=='Na':
        Z.append(11)
Z=np.array(Z)
print(Z)
```

```python
for z,r,a in zip(Z,R,atoms):
    print(z,r,a)
```

```python
# numpy.savez存储多个数组;
# array_dict={'R':R,'N':N,'Z':Z,'id':id_list,'efficiency':efficiency,'smiles':smiles_list}
array_dict={'R':R,'N':N,'Z':Z,'id':id_list,'smiles':smiles_list}
np.savez('3D_canonical_test_lab.npz',**array_dict)
```
